```python
insert_query = f"INSERT INTO sale_tb ({','.join(sale1.columns)}) VALUES ({','.join(['%s'] * len(sale1.columns))})"

for _, row in sale1.iterrows():
    values = tuple(row)
    
    # Check if foreign key value exists in parent table
    foreign_key_value = row['customer_id']  # Replace with actual foreign key column name
    cursor.execute("SELECT customer_id FROM customer_tb WHERE customer_id = %s", (foreign_key_value,))
    
    if cursor.fetchone() is None:
        print(f"Skipping row {values} - Foreign key {foreign_key_value} not found in parent_table")
        continue  # Skip inserting this row

    try:
        cursor.execute(insert_query, values)
    except mysql.connector.Error as err:
        print(f"Error inserting row {values}: {err}")

# Commit changes
mydb.commit()
```
